## Insertion Sort


input: A sequence of n numbers (a_1,a_2,a_3)

Output: A permutation( Reordering)(a_1',a_2',a_3') of the input sequence such that a_1 <= a_2 <= a_3

Conceptually sorting a sequence, the input comes in the form of an array with n elements


<div>

</div>

PseudoCode

for j = 2 to A.length

    key = A[j]
    
    insert key into the sorted sequence
    
    i = j - 1
    
    while i>0 and A[i] > key
    
        A[i+1] = A[i]
        
        i = i - 1
        
    A[i + 1] = key
    

Python

```python
A = [4,3,2,10,12,1,5,6]

for j in range(len(A)):
    key = A[j]
    i = j-1
    while i>=0 and A[i] > key:
        A[i+1] = A[i]
        i = i - 1
    A[i+1] = key
    
print(A)
```
